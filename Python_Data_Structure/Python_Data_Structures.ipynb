**Looping through Strings**


```
fruit = 'banana'
index = 0
while index < len(fruit):
  letter = fruit[index]
  print(index, letter)
  index += 1

for letter in fruit:
  print(letter)
```

**Looping and Counting**

```
fruit = 'banana'
count = 0
for letter in fruit:
  if letter == 'a':
    count += 1
print("Number of 'a' in the fruit",fruit,count,".")
```

**Slicing Strings**

```
s = 'Monty Python'
print(s[0:4])  # end is upto but not including the last
print(s[6:7])
print(s[6:20])  # Okay to reference beyond string index
print(s[:2])    # start from the beginning
print(s[8:])    # upto the end
print(s[:])     # include everything
print(s[-1])    # the last index value
```

<h1> <strong> Dictionaries </strong> </h1>

**List** is a collection of an ordered values <br>
**Dictionaries** are the mess. They have the labels but not order. <br>
Key-Value <br>
**mutable**

```
purse = dict()    # constructor to make empty dictionary
purse['money'] = 12   # in list we would generally use integer numbers and index
purse['candy'] = 3    # while in dictionary we index using labels.
purse['tissues'] = 75
print(purse)
print(purse['candy'])
purse['candy'] += 2
print(purse)
```

**When we want to count how many times a name appears**

```
counts = dict()
names = ['csev', 'cwen', 'csev','zqian','cwen']
for name in names:
  if name not in counts:
    counts[name] = 1
  else:
    counts[name] += 1
print(counts)
```

**The get method for Dictionaries**

```
# two examples that does same thing

if name in counts:
  x = counts[name]    # grab the values and assign it to x
else:
  x = 0

# one line for same thing

x = counts.get(name, 0)   # 0 in there means default is zero if no such key exists
```

```
# Yet another intelligent code

counts = dict()
names = ['csev', 'cwen','csev', 'zqian', 'cwen']
for name in names:
  counts[name] = counts.get(name, 0) + 1
print(counts)
```

**Counting Pattern**

```
counts = dict()
print('Enter a line of text: ')
line = input('')
words = line.split()    # split gives us the list of the words
print('Words: ',words)
print('Counting...')
for word in words:
  counts[word] = counts.get(word,0) + 1
print('Counts', counts)
```

**iterating both keys and values using for loop in dictionary**

**Finding biggest word and biggest count in the file**

```
name = input('Enter a file name: ')
handle = open(name)

counts = dict()
for line in handle:
  words = line.split()
  for word in words:
    counts[word] = counts.get(word,0) + 1

bigcount = None
bigword = None
for word, count in counts.items():
  if bigcount is None or count > bigcount:
    bigcount = count
    bigword = word
print(bigword, bigcount)
```

**Sorting dictionary with Tuple**

```
d = {'a':10, 'b':1, 'c':22}
dict_to_tuples = d.items()
sort_tuple = sorted(dict_to_tuples)
print(sort_tuple)
```

**sort by values instead of key**

```
c = {'a':10, 'b':1, 'c':22}
tmp = list()
for k, v in c.items():
  tmp.append((v,k))

tmp = sorted(tmp, reverse=True)
print(tmp)
```

```
# one of the assignments completed
name = input("Enter file:")
if len(name) < 1 : name = "mbox-short.txt"
handle = open(name)

dic = dict()
for line in handle:
    if line.startswith('From '):
        words = line.split()
        time = words[-2]
        time_split = time.split(':')
        hours = time_split[0]
        dic[hours] = dic.get(hours,0) + 1
        
for k, v in sorted(dic.items()):
    print(k,v)
```

**Find the top 10 most common words**

```
fhand = open('romeo.txt')
counts = dict()
for line in fhand:
  words = line.split()
  for word in words:
    counts[word] = counts.get(word,0) + 1

lst = list()
for key, value in counts.items():
  newtup = (value, key)
  lst.append(newtup)
lst = sorted(lst, reverse=True)

for val, key in lst[:10]:
  print(key, value)
```
